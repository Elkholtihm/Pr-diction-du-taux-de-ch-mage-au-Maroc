# Prediction du taux de chomage au Maroc

##### L'objectif fondamental de ce projet est d'analyser les différents facteurs influençant le taux de chômage au Maroc, en utilisant des données historiques couvrant la période de 1999 à 2022. En se basant sur ces analyses, le projet vise à créer des modèles prédictifs du taux de chômage au Maroc pour les années futures. L'analyse sera segmentée par genre, distinguant entre diplômés et non-diplômés, et prendra également en compte différentes tranches d'âge.

## À propos du dataset

##### En raison du manque d'un ensemble de données complet contenant toutes les variables nécessaires, nous nous sommes basés sur plusieurs ensembles de données provenant du Haut-Commissariat au Plan (HCP). L'ensemble de données comprend le taux de chômage au Maroc de 1999 à 2022, selon le sex, le diplôme et les tranches d'âge entre 15-24 ans, 25-34 ans et 35-44 ans. Nous observons des tendances spécifiques à certaines années, telles que 2020, en ce qui concerne le taux de chômage, ce qui nécessite l'ajout d'autres facteurs justifiant ces tendances, tels que la situation sanitaire et le taux d'investissement.

### Standardisation des données

```python
# Sélection des caractéristiques à standardiser
features = ['Années', 'sexe', 'investissement', 'effet sanitaire']

# Création d'une instance de StandardScaler
scaler = StandardScaler()

# Standardisation des caractéristiques sélectionnées dans le DataFrame 'data_final'
data_final[features] = scaler.fit_transform(data_final[features])
```

```python
data_final.head()
```

### Division des données en ensembles d'entraînement et de test

```python
# Définition des caractéristiques (features) et de la variable cible (target)
target = ['sans d','d moyenne','d superieur', '15-24','25-34','35-44','plus de 45']

# Division des données en ensembles d'entraînement et de test (80% entraînement, 20% test)
x_train, x_test, y_train, y_test = train_test_split(data_final[features], data_final[target], test_size=0.2, random_state=5)
```

## Choix de modèle

#### Les algorithmes utilisés sont le Decision Tree Regressor, le Random Forest Regressor et le Gradient Boosting Regressor pour prédire le taux de chômage au Maroc, en raison des raisons suivantes :
#### * Problème de régression : Les données que nous traitons sont continues ou proches les unes des autres, ce qui indique un problème de régression plutôt qu'un problème de classification. Les modèles de régression sont appropriés pour prédire des valeurs continues.
#### * Non-linéarité et complexité des données : D'après la matrice de corrélation, il n'y a pas de relation linéaire entre la majorité des colonnes de notre ensemble de données. L'utilisation des modèles de régression linéaire n'a pas donné des précisions satisfaisantes, et la nature complexe des données, ainsi que le fait que la cible comprend plus d'une colonne, justifient l'utilisation de modèles tels que les arbres de décision, les forêts aléatoires et les boosters. Ces modèles sont robustes et performants, capables de capturer des relations complexes et de fournir des prédictions précises.
#### * MultiOutputRegressor : La cible comprenant plusieurs colonnes nécessite l'utilisation de la fonction MultiOutputRegressor. Cela permet d'adapter le modèle à plusieurs sorties simultanément, ce qui est approprié dans le cas où l'on essaie de prédire plusieurs variables cibles. (Cependant, l'arbre de décision et le Random Forest permettent également de prédire plus d'une colonne cible.)

## Construction et validaion des modeles:

### Decision Tree Regressor

#### Optimisation des hyperparametres

```python
# Définition des hyperparamètres à optimiser
decision_tree =DecisionTreeRegressor(random_state=5)
param_grid = {
    'max_depth': [k for k in range(1,100,10)],
    'min_samples_split': [k for k in range(2,6)],
    'min_samples_leaf': [k for k in range(1,5)]
}

# utilise tous les cœurs disponibles
grid_search = GridSearchCV(decision_tree, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(x_train, y_train)

# Récupération des meilleurs hyperparamètres
print('Les meilleurs paramètres sont:', grid_search.best_params_)

# Récupération du meilleur modèle
model1 = grid_search.best_estimator_
```

#### Prédiction avec le meilleur modèle.

```python
# Faire des prédictions avec le meilleur modèle sur les ensembles d'entraînement et de test
y_train_pred = model1.predict(x_train)
y_test_pred = model1.predict(x_test)
```

#### Evaluation

```python
# Évaluation de la performance en utilisant le coefficient de détermination R²
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
print('Train R² score = ', train_r2)
print('Test R² score = ', test_r2)
```

```python
# Évaluation de la performance en utilisant l'erreur quadratique moyenne (MSE)
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
print('Train MSE = ', train_mse)
print('Test MSE = ', test_mse)
```

```python
# Évaluation de la performance en utilisant l'erreur absolue moyenne (MAE)
train_mae = mean_absolute_error(y_train, y_train_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
print('Train MAE = ', train_mse)
print('Test MAE = ', test_mse)
```

```python
# Évaluation de la performance en utilisant le racine de l'erreur quadratique moyenne (RMSE) 
train_rmse = np.sqrt(train_mse)
test_rmse = np.sqrt(test_mse)
print('Train RMSE = ', train_rmse)
print('Test RMSE = ', test_rmse)
```

```python
# Conversion array en DataFrame
y_test_pred = pd.DataFrame(y_test_pred, columns=['sans d','d moyenne','d superieur', '15-24','25-34','35-44','plus de 45'])
```

```python
# Définir la taille de la figure
plt.figure(figsize=(10, 10))

# Compteur pour la position du sous-graphique
y = 1

# Parcourir chaque variable cible
for i in target: 
    # Créer un sous-graphique dans une grille 3x3
    plt.subplot(3, 3, y)

    # Tracer les valeurs prédites en bleu
    ax = sns.kdeplot(y_test_pred[i], color='blue', label='Prédit')

    # Tracer les valeurs réelles en rouge
    sns.kdeplot(y_test[i], color='red', ax=ax, label='Réel (Réel)')
    
    # Ajouter la légende
    plt.legend()

    # Incrémenter la position du sous-graphique
    y += 1

# Afficher tous les sous-graphiques ensemble
plt.show()

# Ajouter une disposition serrée pour éviter les chevauchements
plt.tight_layout()
```

###### Les valeurs prédites (en bleu) sont proches des valeurs réelles (en rouge) ce qui indique que le modèle 'Decision Tree Regressor' est bien ajusté.

### Random Forest Regressor

#### Optimisation des hyperparametres

```python
%%time
# Définition des hyperparamètres à optimiser
param = {
    'estimator__n_estimators': [100, 500, 1000],
    'estimator__max_depth': [None, 5, 10, 15],
    'estimator__min_samples_split': [2, 5, 10],
    'estimator__min_samples_leaf': [1, 2, 4],
}

# utilise tous les cœurs disponibles
with parallel_backend('loky'):
    grid = GridSearchCV(MultiOutputRegressor(estimator=RandomForestRegressor(random_state=42)), param_grid=param, scoring='r2', cv=5)

    grid.fit(x_train, y_train)
# Récupération des meilleurs hyperparamètres
best_params = grid.best_params_
print("Meilleurs hyperparamètres :", best_params)

# Récupération du meilleur modèle
model2 = grid.best_estimator_
```

#### Prédiction avec le meilleur modèle.

```python
# Faire des prédictions avec le meilleur modèle sur les ensembles d'entraînement et de test
y_train_pred = model2.predict(x_train)
y_test_pred = model2.predict(x_test)
```

#### Evaluation

```python
# Évaluation de la performance en utilisant le coefficient de détermination R²
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
print('Train R² score = ', train_r2)
print('Test R² score = ', test_r2)
```

```python
# Évaluation de la performance en utilisant l'erreur quadratique moyenne (MSE)
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
print('Train MSE = ', train_mse)
print('Test MSE = ', test_mse)
```

```python
# Évaluation de la performance en utilisant l'erreur absolue moyenne (MAE)
train_mae = mean_absolute_error(y_train, y_train_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
print('Train MAE = ', train_mse)
print('Test MAE = ', test_mse)
```

```python
# Évaluation de la performance en utilisant le racine de l'erreur quadratique moyenne (RMSE) 
train_rmse = np.sqrt(train_mse)
test_rmse = np.sqrt(test_mse)
print('Train RMSE = ', train_rmse)
print('Test RMSE = ', test_rmse)
```

### Gradient Boosting Regressor

#### Optimisation des hyperparamètres.

```python
%%time
# Définition des hyperparamètres à optimiser
param = {
    'estimator__n_estimators': [100, 500, 1000],
    'estimator__learning_rate': [0.01, 0.1, 0.2],
    'estimator__max_depth': [3, 5, 7],
    'estimator__min_samples_split': [2, 5, 10],
    'estimator__min_samples_leaf': [1, 2, 4],
    'estimator__subsample': [0.8, 0.9, 1.0],
    'estimator__random_state': [42, 567]
}

# utilise tous les cœurs disponibles
with parallel_backend('loky'):
    grid = GridSearchCV(MultiOutputRegressor(estimator=GradientBoostingRegressor()), param_grid=param, scoring='r2', cv=5)

    grid.fit(x_train, y_train)

best_params = grid.best_params_
print("Meilleurs hyperparamètres :", best_params)
model3 = grid.best_estimator_
```

#### Prédiction avec le meilleur modèle.

```python
# Faire des prédictions avec le meilleur modèle sur les ensembles d'entraînement et de test
y_train_pred = model3.predict(x_train)
y_test_pred = model3.predict(x_test)
```

#### Evaluation

```python
# Évaluation de la performance en utilisant le coefficient de détermination R²
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
print('Train R² score = ', train_r2)
print('Test R² score = ', test_r2)
```

```python
# Évaluation de la performance en utilisant l'erreur quadratique moyenne (MSE)
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
print('Train MSE = ', train_mse)
print('Test MSE = ', test_mse)
```

```python
# Évaluation de la performance en utilisant l'erreur absolue moyenne (MAE)
train_mae = mean_absolute_error(y_train, y_train_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
print('Train MAE = ', train_mse)
print('Test MAE = ', test_mse)
```

```python
# Évaluation de la performance en utilisant le racine de l'erreur quadratique moyenne (RMSE) 
train_rmse = np.sqrt(train_mse)
test_rmse = np.sqrt(test_mse)
print('Train RMSE = ', train_rmse)
print('Test RMSE = ', test_rmse)
```
